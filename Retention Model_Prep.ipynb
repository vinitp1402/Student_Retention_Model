# Subsetting and merging to set Target variable

```python
# Getting unique combination of Enrolment Year, Student and School
df1 = stu_ds[['Enrolment_Year','Student_No','School']].drop_duplicates()
```

```python
df_2015 = df1[df1['Enrolment_Year'] == 2015]
df_2016 = df1[df1['Enrolment_Year'] == 2016]
df_2017 = df1[df1['Enrolment_Year'] == 2017]
df_2018 = df1[df1['Enrolment_Year'] == 2018]
df_2019 = df1[df1['Enrolment_Year'] == 2019]
```

```python
df_15_16 = pd.merge(df_2015, df_2016, on=('Student_No', 'School'), how = 'left')
df_16_17 = pd.merge(df_2016, df_2017, on=('Student_No', 'School'), how = 'left')
df_17_18 = pd.merge(df_2017, df_2018, on=('Student_No', 'School'), how = 'left')
df_18_19 = pd.merge(df_2018, df_2019, on=('Student_No', 'School'), how = 'left')
df_15_18 = pd.concat([df_15_16, df_16_17, df_17_18, df_18_19])
```

# Setting values for Retained

```python
df_15_18['Retained'] = 'Not Set'
# If Enrolment_Year_y is not NaN, it means the student did continue the following year
# NaN is introduced as a result of the left join
df_15_18.loc[df_15_18['Enrolment_Year_y'].isnull(), 'Retained'] = 0
df_15_18.loc[df_15_18['Enrolment_Year_y'].notnull(), 'Retained'] = 1

# Rename the column names
df_15_18 = df_15_18.rename(columns={"Enrolment_Year_x": "Enrolment_Year"})
df_15_18.head(10)
```

# Handling missing / garbage values

```python
stu_ds["ATAR_Band"].fillna("NA", inplace = True)

# using mean as it should not affect the mean of whole dataset
#stu_ds["WAM Score"].fillna(stu_ds['WAM Score'].mean(), inplace = True)
# Not using WAM Score as 98% of the values are null

stu_ds["Course_Commencing_Status"].fillna("NA", inplace = True)
stu_ds["Age"].fillna("NA", inplace = True)
```

```python
# The column ATAR_Band contains some records as 'No ATAR'
# Replacing them with NA
# This brings in uniformity merging No ATAR and NA values. They both mean information is not available!
stu_ds.replace(to_replace ="No ATAR",  
                            value ="NA", inplace = True) 
```

# Prepare a df with agg statistics
1. mean for SFS, SFT, NPS
1. count for Subjects
1. count for Course
1. count for Course Location
1. count for unique subjects taken
1. No of years with the school



```python
stu_agg_ds = stu_ds.groupby(['Enrolment_Year','Student_No','School'], as_index=False).agg(
    {
        'SFS':"mean",
        'SFT':"mean",
       	'NPS':"mean",
        'Subject':  "count", # Subjects could be repeated
        'Course':   "nunique",
        'Course_Location_Name': "nunique"
    }
     )
stu_agg_ds.head(5)
```

### 2,3,4. Count for Subjects, Course and Course Location

```python
stu_cnt_ds = stu_ds.groupby(['Enrolment_Year','Student_No','School'], as_index=True).agg({
    'Subject':  "count", # Subjects could be repeated
    'Course':   "nunique",
    'Course_Location_Name': "nunique"
}).reset_index()
stu_cnt_ds = stu_cnt_ds.rename(columns={"Subject": "Subjects_Count",
                                            "Course": "Course_Uniq",
                                            "Course_Location_Name": "Course_Loc_Uniq"})
stu_cnt_ds.head(20)
```

### 5. Count for Unique subjects taken

```python
stu_cnt_ds_1 = stu_ds.groupby(['Enrolment_Year','Student_No','School'], as_index=True).agg({
    'Subject':'nunique'
}).reset_index()
stu_cnt_ds_1 = stu_cnt_ds_1.rename(columns={"Subject": "Subjects_Uniq"})
stu_cnt_ds_1.head()
```

# Merge the two dataframes that contain the aggregated information

```python
stu_agg2 = pd.concat([
                      stu_agg_ds.iloc[:,0:9], 
                      #stu_cnt_ds.iloc[:,3:],
                      stu_cnt_ds_1.iloc[:,3:]],
                      axis = 1)
stu_agg2.head(5)
```

```python
stu_year_cnt = df1.groupby(['Student_No', 'School'], as_index=True).agg({
    "Enrolment_Year": 'count'
}).reset_index()
stu_year_cnt = stu_year_cnt.rename(columns={"Enrolment_Year": "No of Years"})
```

```python
stu_agg2 = pd.merge(stu_agg2, stu_year_cnt, on=('Student_No',
                                     'School'),
         how = 'inner')
```

```python
stu_agg2.head(5)
```

**Need to map the target variable to the original dataset stu_ds. Then stu_ds can be turned to a wide data frame.**

```python
stu_ds_merged = pd.merge(stu_agg2, df_15_18, on=('Enrolment_Year','Student_No', 'School'), how = 'inner')
stu_ds_merged.head(50)
```

## An Observation for 'ATAR_Band', 'Age', 'Course_Category_Group'
* **These values stay the same for the combination year-student-school. So Just get the first value when grouped by the key combination**
* **Then combine this information with agg results**

```python
stu_ds_cat_condensed = stu_ds.groupby(['Enrolment_Year','Student_No','School'], as_index=True)['ATAR_Band', 'Age', 'Course_Category_Group'].first().reset_index()
stu_ds_cat_condensed.head(10)
```

Drop the values that coontain Enrolment_Year = 2019 as these are current year students and do not have the target variable

```python
stu_ds_merged = pd.merge(stu_ds_cat_condensed, stu_ds_merged, on=('Enrolment_Year','Student_No', 'School'), how = 'inner')
```

```python
stu_ds_merged.sort_values(by=['Student_No', 'Enrolment_Year', 'School'])
```

**We can calculate if the student repeated subjects. That can be calculated as Subjects_Count - Subjects_Uniq**

```python
stu_ds_merged["Subjects_Repeated"] = stu_ds_merged['Subject']-stu_ds_merged['Subjects_Uniq']
stu_ds_merged = stu_ds_merged.drop(columns=['Subject', 'Enrolment_Year_y'])
stu_ds_merged.head(10)
```
